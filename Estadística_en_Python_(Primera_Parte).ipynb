```python
import pandas as pd
import io

# Cargar el CSV desde Google Colab
from google.colab import files
uploaded = files.upload()

# Imprimir las claves disponibles
print("Claves disponibles:", uploaded.keys())

# Obtener el nombre del archivo (clave)
file_name = list(uploaded.keys())[0]  # Obtener el primer (y único) elemento de las claves
print("Nombre del archivo:", file_name)

# Leer el CSV
# Decodificar los bytes a una cadena antes de usar io.StringIO
df = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Resto del código...


# Asegurarse de que la columna 'casos' esté en formato numérico (eliminar comas)
df['casos'] = df['casos'].str.replace(',', '').astype(int)

# 1. Distribución de frecuencias. Datos Cualitativos.
frecuencia_cualitativos = df['pais'].value_counts()
print("1. Distribución de frecuencias. Datos Cualitativos:")
print(frecuencia_cualitativos)

# 2. Distribución de frecuencias. Datos Cuantitativos.
frecuencia_cuantitativos = df['casos'].value_counts().sort_index()
print("\n2. Distribución de frecuencias. Datos Cuantitativos:")
print(frecuencia_cuantitativos)

# 3. Distribuciones de frecuencia relativa y de frecuencia porcentual. Datos Cualitativos.
frecuencia_relativa_cualitativos = df['pais'].value_counts(normalize=True)
frecuencia_porcentual_cualitativos = df['pais'].value_counts(normalize=True) * 100
print("\n3. Distribuciones de frecuencia relativa y de frecuencia porcentual. Datos Cualitativos:")
print("Frecuencia Relativa:")
print(frecuencia_relativa_cualitativos)
print("\nFrecuencia Porcentual:")
print(frecuencia_porcentual_cualitativos)

# 4. Distribuciones de frecuencia relativa y de frecuencia porcentual. Datos Cuantitativos.
frecuencia_relativa_cuantitativos = df['casos'].value_counts(normalize=True).sort_index()
frecuencia_porcentual_cuantitativos = df['casos'].value_counts(normalize=True).sort_index() * 100
print("\n4. Distribuciones de frecuencia relativa y de frecuencia porcentual. Datos Cuantitativos:")
print("Frecuencia Relativa:")
print(frecuencia_relativa_cuantitativos)
print("\nFrecuencia Porcentual:")
print(frecuencia_porcentual_cuantitativos)

```
